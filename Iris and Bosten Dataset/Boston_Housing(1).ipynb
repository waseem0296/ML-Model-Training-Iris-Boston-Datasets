```python
import pandas as pd
import numpy as np

# Define column names
columns = [
    "crime_rate", "residential_zone", "industrial_zone", "near_river",
    "nitric_oxide", "avg_rooms", "old_houses", "distance_to_jobs",
    "highway_access", "property_tax", "student_teacher_ratio",
    "black_population", "lower_income", "house_price"
]

# Load the dataset with correct column names
df = pd.read_csv('housing.csv', names=columns, skiprows=1, delim_whitespace=True)

# Display the first few rows
print(df.head())

df.info()
df.isnull().sum()
print("\nChecking for duplication")
df.duplicated().sum()
#df = df.drop_duplicates()
print("\nGetting summary statistics")
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()



df.hist(figsize=(12, 10), bins=20, color='skyblue', edgecolor='black')
plt.suptitle("Boston Housing Data Distribution", fontsize=16)
plt.show()



plt.figure(figsize=(10, 12))


sns.boxplot(data=df, palette="pastel", orient="h")

plt.title("Boxplot of Dataset Features", fontsize=14)
plt.xlabel("Values", fontsize=12)
plt.ylabel("Features", fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 4))
plt.scatter(df['avg_rooms'], df['house_price'], alpha=0.5)
plt.xlabel('Average Number of Rooms')
plt.ylabel('House Price')
plt.title('Relationship between Rooms and Price')
plt.show()
```

```python
from scipy import stats

z_scores = stats.zscore(df)
outliers = (abs(z_scores) > 3).sum()
print("Number of outliers per column:", outliers)


df.skew()

#right skew
df['crime_rate'] = np.log1p(df['crime_rate'])  # Log Transformation
df['residential_zone'] = np.log1p(df['residential_zone'])
df['near_river'] = np.log1p(df['near_river'])
df['lower_income'] = np.log1p(df['lower_income'])
df['house_price'] = np.log1p(df['house_price'])

df['distance_to_jobs'] = np.sqrt(df['distance_to_jobs'])  # Square Root Transformation
df['highway_access'] = np.sqrt(df['highway_access'])

df['property_tax'] = np.cbrt(df['property_tax'])  # Cube Root Transformation

#left skew
df['old_houses'] = df['old_houses'] ** 2  # Square Transformation
df['student_teacher_ratio'] = df['student_teacher_ratio'] ** 3  # Cube Transformation
df['black_population'] = np.cbrt(df['black_population'])
print(df.skew())
df.hist(figsize=(12, 10), bins=20, color='skyblue', edgecolor='black')
```

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
df_iris['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
print(df_iris.columns)
```

```python
for species in iris.target_names:
    species_data = df_iris[df_iris['species'] == species]
    print(f"\nStatistics for {species}:")

    # Exclude 'species' column from calculations
    numeric_data = species_data.drop(columns=['species'])

    print("Mean:", numeric_data.mean())
    print("Median:", numeric_data.median())
    print("Standard Deviation:", numeric_data.std())
    print("Range:", numeric_data.max() - numeric_data.min())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
# Density plots with species comparison
plt.figure(figsize=(12, 8))
for i, feature in enumerate(iris.feature_names):
    plt.subplot(2, 2, i + 1)  # Use 'i' instead of index() for safety
    for species in iris.target_names:
        species_data = df_iris[df_iris['species'] == species][feature]
        sns.kdeplot(data=species_data, label=species, fill=True)
    plt.title(f'Density Plot of {feature}')
    plt.legend()

plt.tight_layout()  # Adjust layout for better visibility
plt.show()

# Violin plots for distribution comparison
plt.figure(figsize=(15, 6))
for i, feature in enumerate(iris.feature_names):
    plt.subplot(2, 2, i + 1)
    sns.violinplot(x='species', y=feature, data=df_iris)
    plt.title(f'Violin Plot of {feature}')

plt.tight_layout()
plt.show()
```

```python

# 1. Enhanced Pair Plot with Density
sns.pairplot(df_iris,
             hue='species',
             diag_kind='kde',
             plot_kws={'alpha': 0.6})

# 2. Separate Feature Analysis
for feature in iris.feature_names:
    plt.figure(figsize=(12,6))

    # Histogram with density overlay
    plt.subplot(1,2,1)
    sns.histplot(data=df_iris, x=feature, hue='species', stat='density',
common_norm=False, alpha=0.4)

    # Box plot
    plt.subplot(1,2,2)
    sns.boxplot(data=df_iris, x='species', y=feature)
    plt.xticks(rotation=45)

    plt.tight_layout()
```
